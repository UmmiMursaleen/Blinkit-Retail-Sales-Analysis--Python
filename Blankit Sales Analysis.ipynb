## Ways to detect Outlier

```python
# Using Box plot
plt.figure(figsize=(10,4))
sns.boxplot(x=blankit['Sales'])
plt.show()

#As visual shows that there is no outlier in our dataset
```

```python
# Using Z score

from scipy import stats 
zscore=np.abs(stats.zscore(blankit['Sales']))
blankit[zscore>3]
```

Using Summary Statistics IQR

```python
q1 = blankit['Sales'].quantile(0.25)
q3 = blankit['Sales'].quantile(0.75)
iqr = q3-q1

lowerbond = q1 - 1.5 * iqr
upperbond = q3 + 1.5 * iqr

outlier=blankit[(blankit['Sales']<lowerbond) | (blankit['Sales']>upperbond)]
print(outlier)
```

```python
#Top 10 Product Sales

top10=blankit.groupby('Item Type').agg({'Sales':'sum'}).sort_values('Sales',ascending=False)
```

```python
plt.figure(figsize=(10,6))
bars=plt.bar(top10.index.astype(str),top10['Sales'])
plt.xticks(rotation=-90)

#Data Label

for bar in bars:
    height=bar.get_height()
    plt.text(bar.get_x()+bar.get_width()/2,
             height+3,
             f'{height:,.0f}',
             ha='center',va='bottom',
             fontsize=8
            )

plt.tight_layout()
plt.show()
```

```python
itemfat=blankit.groupby('Item Fat Content').agg({'Sales':'sum'})


plt.figure(figsize=(4,4))
plt.pie(itemfat['Sales'],labels=itemfat.index.astype(str),autopct="%1.1f%%",startangle=-90)
plt.title('Sales of Item Fat Content',weight='bold')
plt.show()
```

```python
outletsize=blankit.groupby('Outlet Size').agg({'Sales':'sum'})
plt.figure(figsize=(4,4))
plt.pie(outletsize['Sales'],labels=outletsize.index.astype(str),autopct="%1.1f%%",startangle=-90)
plt.tight_layout()
plt.show()
```

```python
rating=blankit.groupby('Outlet Identifier').agg({'Rating':'mean'}).sort_values('Rating',ascending=False)

plt.figure(figsize=(8,6))
bars = plt.barh(rating.index.astype(str), rating['Rating'], color='steelblue')
plt.xlabel('Average Rating', fontsize=12)
plt.title('Average Rating by Outlet', fontsize=14, weight='bold')
plt.gca().invert_yaxis()  # highest on top

# Add data labels
for i, v in enumerate(rating['Rating']):
    plt.text(v + 0.05, i, f'{v:.2f}', va='center', fontsize=9)

plt.tight_layout()
plt.show()
```
